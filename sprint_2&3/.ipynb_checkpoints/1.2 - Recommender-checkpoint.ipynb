## Get frontpage data

```python
frontpage_url = "https://www.swissinfo.ch/webservice/swi-eng-2.0/overview"
r = requests.get(frontpage_url).json()
```

```python
items = []
for item in r["page"]["item"][0]["content"]:
    try:
        items.append({"id": item["id"], "title": item["title"], "link": item["canonical"], 
                 "image": "https://www.swissinfo.ch%s" % item["smallImage"], 
                 "date": item["dateLoc"], "description": item["text"],
                     "tag": item["subjectTags"][0]["title"]}) 
    except:
        pass
items[0]
```

## Setup and Reset Database

```python
from recombee_api_client.api_client import RecombeeClient
from recombee_api_client.api_requests import AddItemProperty, SetItemValues, AddPurchase
from recombee_api_client.api_requests import RecommendItemsToItem, Batch, ResetDatabase
import random

NUM = len(items)
PROBABILITY_PURCHASED = 0.1

client = RecombeeClient('swissinfo', '4ndNA3jjJr3gRVJj7jekF1qVhAGXXSJP58kHGd3RgIfX85a5CHGF99Pbbx5G3ceD')

#Clear the entire database
client.send(ResetDatabase())

# Add properties of items
client.send(AddItemProperty('title', 'string'))
client.send(AddItemProperty('link', 'string'))
client.send(AddItemProperty('image', 'image'))
client.send(AddItemProperty('date', 'string'))
client.send(AddItemProperty('tag', 'string'))
client.send(AddItemProperty('description', 'string'))
```

```python


# Prepare requests for setting a catalog of computers
requests = [SetItemValues(
    "item-%s" % item["id"], #itemId
    #values:
    {
      'title': item["title"],
      'link': item["link"],
      'image': item["image"],
      "date": item["date"],
       "tag": item["tag"],
      'description': item["description"]
    },
    cascade_create=True   # Use cascadeCreate for creating item
                          # with given itemId if it doesn't exist
  ) for item in items]


# Send catalog to the recommender system
client.send(Batch(requests))
```

```python
requests = []
users = ["user-%s" % i for i in range(NUM)]

for item in items:
    #Use cascadeCreate to create unexisting users
    purchasing_users = [user_id for user_id in users if random.random() < PROBABILITY_PURCHASED]
    requests += [AddPurchase(user_id, "item-%s" % item["id"], cascade_create=True) for user_id in purchasing_users]

# Send purchases to the recommender system
client.send(Batch(requests))
```
