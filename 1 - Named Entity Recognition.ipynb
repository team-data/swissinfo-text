# Named Entities Recognition

```python
from nltk import ne_chunk, pos_tag, word_tokenize
from nltk.tree import Tree

def get_continuous_chunks(text):
    chunked = ne_chunk(pos_tag(word_tokenize(text)))
    prev = None
    continuous_chunk = []
    current_chunk = []

    for i in chunked:
        if type(i) == Tree:
            current_chunk.append(" ".join([token for token, pos in i.leaves()]))
        elif current_chunk:
            named_entity = " ".join(current_chunk)
            if named_entity not in continuous_chunk:
                continuous_chunk.append(named_entity)
                current_chunk = []
        else:
            continue
    if current_chunk:
        named_entity = " ".join(current_chunk)
        if named_entity not in continuous_chunk:
            continuous_chunk.append(named_entity)
            current_chunk = []
    return continuous_chunk
```

```python
r = get_continuous_chunks("Johan Meier kommt aus Deutschland und arbeitet bei der Forschungsabteilung von Microsoft")
r
```

```python
result = get_continuous_chunks(text)
result[0:30]
```

## Lets do it for 100 Frontpage items

```python
# Get 100 Frontpage Items
frontpage_url = "https://www.swissinfo.ch/webservice/swi-eng-2.0/overview"
r = requests.get(frontpage_url).json()
items = []
urls = []
for item in r["page"]["item"][0]["content"]:
    try:
        tmp = requests.get("https://www.swissinfo.ch/webservice/swi-eng-2.0/detail%s" % item["url"])
        text = BeautifulSoup(tmp.json()["htmldetail"], "lxml").get_text()
        items.append(text)
        urls.append(item["canonical"])
        print("Done: %s" % item["url"])
    except:
        pass
```

```python
result = []
for item in items:
    result += get_continuous_chunks(item)
```

```python
stopwords = ["Array", "Click","HEAD","HTML","Email","CSS","Society","Array jQuery", "Swiss Switzerland", "Switzerland Switzerland", "KEYSTONE","Switzerland MailChimp","MailChimp", "Switzerland Swiss"]
result_filtered = [word for word in result if word not in stopwords]
df = pd.DataFrame(result_filtered,columns=["entities"])
df['entities'].value_counts()[0:30].plot(kind="bar",figsize=(15, 6))
```
